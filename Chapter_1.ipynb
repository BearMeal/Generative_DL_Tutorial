## 생성 모델개요
- 생성 모델링은 학습 데이터셋에 확률적으로 있을법한 특성(feature)를 가진 새로운 샘플을 만들어낸다.
- 어떤 데이터가 어떤 라벨링을 가졌는디 판별하는 판별 모델링(discriminative)과는 다르다.
- 특성은 이미지, 단어, 문자 그룹등 일 수 있다.
- 생성 모델은 probabilistic 해야 한다.
- 조건부 생성모델: 여러과일이 있는 데이터셋에서 사과 이미지를 생성하는 모델
- 현실을 인지하는것이 생성 모델처럼 확률적일수있다
- 가장 좋은 생성 모델은 인간이다. 따라서 인간과 견줄만한 지능이 있는 기계가 나온다면 생성 모델이 솔루션의 일부가 될것이다. 사람 뇌를 학습한다면 그것 또한 인공지능인가

### 간단한 생성모델개념
- 2차원 공간에 존재하는 점들의 P(x)를 구한다.
- 새로운 데이터 P(model) 일때를 본다면 새로운 점의 위치가 가능한 영역인가 알수있다.
- 고차원 데이터의 표현(represetation,sen이 아닌건 오타가 아니다.)
- 예를 들어 금발에 안경낀 사람을 찾는다고 픽셀1~픽셀x 까지 색상을 설명할순없다.
따라서 고차원의 표본공간을 직접모델링하는것이 아니라. 잠재공간(latent space)을 사용해 훈련세트의 각 샘플을 원본 공간의 포인트에 매핑 합니다. 

### 핵심 이론: 확률
- 표본공간(sample space): 샘플 x가 가질수 있는 모든 값의 집합
  - 생성분포에 x가 포함된다면 생성모델링에서 샘플링한 샘플이 될수 있다.
- 확률밀도함수(probability density function): 포인트 x를 0과 1사이의 숫자로 매핑하는 함수p(x)이다. 표본공간에 있는 모든 포인트에 대해 밀도함수를 적분했을때 1이 되어야 잘정의된 확률 분포이다.
  - 2차원 세계지도 예제에서 생성밀도 함수는 생성분포밖은 0 안은 일정이다. 따라서 전체 표본 공간에 대해 밀도함수를 적분하면 1이된다.
- 모수 모델링(parametric modeling): 안정적인 $p_{model}(x)$를 찾는데 사용할 기법
  - 유한한 개수의 파라미터 $\theta$를 사용해기술할수 있는 밀도함수 $p_{\theta}(x)$의 한종류이다.
- 가능도(likelyhood): 파라미터 집합 $\theta$의 가능도 $L(\theta\vert x)$ 는 관측된 포인트 x가 주어졌을때 $\theta$의 타당성을 측정하는 함수
$$L(\theta\vert x) = p_0(x)$$
- 최대 가능도추정(maximum likelyhood estimate,MLE): $\hat{\theta}$ 을 추정할수 있는 기법.$\hat{\theta}$ 는 관측된 데이터X를 가장 잘 설명하는 밀도함수$p_\theta(x)$의 파라미터 집합 $\theta$이다. $$\hat{\theta} = argmax\lambda(\theta\vert X)$$
